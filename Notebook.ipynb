## Sprint 1 - Definição do Produto

### 1. Definição da Empresa Fictícia
- **Nome:** FoodHunter
- **Missão:** Ajudar pessoas a decidir rapidamente onde comer usando IA personalizada e contextual.
- **Visão:** Ser a IA referência em recomendações gastronômicas locais.
- **Diferencial:** Não é apenas um buscador padrão. É uma IA conversacional que entende preferências implícitas e sugere opções sob medida, integrando dados comportamentais.

### 2. Definição do Problema de Negócio
- **Problema:** Escolher onde comer é um processo frequentemente demorado e confuso, agravado pelo excesso de opções e pela falta de personalização nos aplicativos tradicionais de delivery e mapas.
- **Dores do Usuário:**
  - Indecisão constante ("O que vamos comer hoje?").
  - Falta de recomendações realmente relevantes ao contexto (clima, humor, restrições).
  - Filtros engessados (apenas por "tipo de cozinha" ou "distância").
  - Tempo perdido comparando opções em múltiplos apps.
- **Solução Proposta:** Uma plataforma baseada em IA que entende linguagem natural, considera o contexto do usuário (localização, horário, perfil comportamental) e realiza recomendações diretas, interativas e inteligentes.

### 3. Levantamento de Requisitos
**Requisitos Funcionais:**
- [RF01] O sistema deve receber perguntas em linguagem natural via chat.
- [RF02] O sistema deve processar a intenção do usuário utilizando NLP/LLM.
- [RF03] O sistema deve classificar o perfil gastronômico do usuário com base em um modelo de Machine Learning treinado.
- [RF04] O sistema deve buscar restaurantes cruzando a localização do usuário com APIs externas.
- [RF05] O sistema deve manter um histórico/banco de preferências do usuário.

**Requisitos Não Funcionais:**
- [RNF01] Tempo de resposta conversacional rápido (idealmente < 2 segundos).
- [RNF02] Alta disponibilidade da API e do modelo de IA.
- [RNF03] Privacidade e segurança dos dados e histórico de preferências do usuário.
- [RNF04] Escalabilidade para suportar múltiplos usuários simultâneos.

## 🏗️ Sprint 2 - Arquitetura e Infraestrutura Base

### 1. Definição da Arquitetura Geral
A arquitetura do **FoodHunter** foi projetada para ser modular, conteinerizada e orientada ao fluxo de dados de um sistema RAG (Retrieval-Augmented Generation). 

**Componentes Principais:**
* **Frontend (UI):** Interface interativa desenvolvida em Streamlit/Gradio para a experiência de chat do usuário.
* **Backend (API):** API RESTful construída em FastAPI para orquestrar as requisições, conectar com o LLM e gerenciar a lógica de negócio.
* **Motor RAG & NLP:** Utilização da biblioteca `sentence-transformers` (PyTorch) para gerar embeddings dinâmicos da pergunta do usuário e realizar a busca semântica (Cosseno de Similaridade).
* **IA Generativa:** Integração com um LLM (ex: Ollama) para ler os contextos encontrados e formular a recomendação em linguagem natural.
* **Data Lake (Medallion):** MinIO (Object Storage compatível com S3) atuando como repositório de arquivos organizados nas camadas Bronze, Silver e Gold (`.parquet`).
* **Banco de Dados:** PostgreSQL para armazenar dados estruturados, como perfis de usuários e histórico de conversas.
* **Infraestrutura:** Docker e Docker Compose para garantir a portabilidade e fácil execução de todos os serviços.

### 2. Setup Inicial (Docker Compose)
O ambiente local é provisionado rapidamente através de um arquivo `docker-compose.yaml`, que sobe os seguintes serviços isolados em rede:
- `foodhunter-api` (Porta 8000)
- `foodhunter-ui` (Porta 8501)
- `minio-storage` (Porta 9000 para API, 9001 para Console)
- `postgres-db` (Porta 5432)

## 🗄️ Sprint 3 - Governança e Medallion Architecture

A gestão dos dados do FoodHunter foi estruturada utilizando a arquitetura Medallion no MinIO, garantindo rastreabilidade e qualidade para o motor de IA. 

A documentação detalhada, as análises exploratórias e os scripts de processamento de cada etapa estão isolados em seus respectivos notebooks:
* **🥉 Camada Bronze (`bronze/bronze.ipynb`):** Ingestão dos dados brutos do dataset Yelp e análise de schema.
* **🥈 Camada Silver (`silver/silver.ipynb`):** Limpeza de JSONs stringificados, tipagem de variáveis e extração de features booleanas (delivery, outdoor seating) e numéricas (preço).
* **🥇 Camada Gold (`gold/gold.ipynb`):** Criação do `gold_score` (balanceamento de notas e volume de reviews) e geração do `rag_context` para consumo do LLM.

## 🧠 Sprint 4 - Modelagem e Motor de Recomendação (RAG)

### 1. Definição do Problema de Machine Learning
O FoodHunter não utiliza algoritmos clássicos de predição linear para recomendar restaurantes. O núcleo da inteligência é um sistema **RAG (Retrieval-Augmented Generation)** baseado em similaridade vetorial.

* **O Modelo Gerador:** Utilizamos NLP profundo através da biblioteca `sentence-transformers` (engine PyTorch) com o modelo `all-MiniLM-L6-v2` para transformar as descrições dos restaurantes da camada Gold em tensores densos (embeddings customizados).
* **O Motor de Busca:** A intenção do usuário é vetorizada em tempo real, e a recomendação é feita calculando a **Similaridade de Cosseno** entre a pergunta e a nossa base de conhecimento.

### 2. Rastreamento de Experimentos (MLflow)
Para garantir a reprodutibilidade metodológica, o MLflow é utilizado para logar:
* O modelo de linguagem utilizado na geração dos embeddings.
* O tempo de processamento da vetorização em lote.
* Os hiperparâmetros da busca (como o número de resultados Top-K retornados para o LLM).
